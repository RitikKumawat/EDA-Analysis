```python
import seaborn as sns
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df['class'] = iris.target
df.columns =  ['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
```

```python
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='sepal_len',y='sepal_wid',data=df,kind='hex')
```

```python
sns.jointplot(x='sepal_len',y='sepal_wid',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='class')
```

```python
sns.countplot(x='class',data=df)
```

```python
sns.set()
fig,axes = plt.subplots(2,2,figsize=(15,15))

fig.subplots_adjust(hspace=0.3)

sns.boxplot(y='sepal_len',x='class',data=df,ax=axes[0,0])
sns.boxplot(y='sepal_wid',x='class',data=df,ax=axes[0,1])
sns.boxplot(y='petal_len',x='class',data=df,ax=axes[1,0])
sns.boxplot(y='petal_wid',x='class',data=df,ax=axes[1,1])
```
